## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
application_df.nunique()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
app_df_clean=application_df.drop(columns=["EIN"])
```

```python
#check value counts of specific columns(STATUS and SPRECIAL CONSIDERATIONS)
app_df_clean['STATUS'].value_counts()
```

```python
app_df_clean['SPECIAL_CONSIDERATIONS'].value_counts()
```

```python
app_df_clean=app_df_clean.drop(columns=['STATUS', 'SPECIAL_CONSIDERATIONS'])
app_df_clean.head()
```

```python
# Look at NAME value counts for binning
name_counts=app_df_clean['NAME'].value_counts()
name_counts
```

```python
name_counts[name_counts<10]
```

```python
# Choose a cutoff value and create a list of NAME to be replaced
# replace if counts are less than 10
# use the variable name `name_to_replace`
names_to_replace = list(name_counts[name_counts <10].index)
# Replace in dataframe
for app in names_to_replace:
    app_df_clean['NAME'] = app_df_clean['NAME'].replace(app,"Other")

# Check to make sure binning was successful
app_df_clean['NAME'].value_counts()
```

```python
# Review APPLICATION_TYPE value counts for binning
app_type_counts = app_df_clean['APPLICATION_TYPE'].value_counts()
app_type_counts
```

```python
# Choose a cutoff value and create a list of application to be replaced
# use the variable name `app_types_to_replace`
app_types_to_replace = list(app_type_counts[app_type_counts < 500].index)


# Replace in dataframe
for cls in app_types_to_replace:
    app_df_clean['APPLICATION_TYPE'] = app_df_clean['APPLICATION_TYPE'].replace(cls,"Other")
    
# Check to make sure binning was successful
app_df_clean['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_counts = app_df_clean['CLASSIFICATION'].value_counts()
class_counts
```

```python
# Choose a cutoff value and create a list of CLASSIFICAION to be replaced
# use the variable name `class_to_replace`
class_to_replace = list(class_counts[class_counts < 1000].index)

# Replace in dataframe
for cls in class_to_replace:
    app_df_clean['CLASSIFICATION'] = app_df_clean['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
app_df_clean['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
app_df_clean = pd.get_dummies(app_df_clean)
app_df_clean.head()
```

```python
# Split our preprocessed data into our features and target data
X = app_df_clean.drop(columns=['IS_SUCCESSFUL']).values
y = app_df_clean['IS_SUCCESSFUL'].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
num_input_features = 263 # X.shape
hidden_nodes_layer1 =  100
hidden_nodes_layer2 = 30
hidden_nodes_layer3 = 10

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=80, activation="relu", input_dim=num_input_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=30, activation="sigmoid"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=8, activation="sigmoid"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=200)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
